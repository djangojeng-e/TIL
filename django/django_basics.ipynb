# Web programming 


naver, google 혹은 유튜브 같은 서비스를 웹 서비스라고 얘기한다. 

World Wid Web 의 줄임말로, www 가 주소에 나와 있는 이유이다. 
web page 와 web site 로 분류가 되는데. 여러개의 웹 페이지가 합쳐져서 하나의 웹 사이트를 만들어 낸다. 

1. 웹 프로그램이란? 
웹 프로그래밍은 (web programming) 은, 컴퓨터 언어, 즉, 프로그래밍 언어, 코드로 이루어져 있는 웹사이트 (혹은 웹페이지) 를 만드는 과정이다. 
웹 프로그램밍을 한다는것은, 웹사이트 혹은 페이지를 만들거나 수정하는 행위이다 

웹 브라우저를 통하여, 이렇게 프로그래밍 된 페이지나 사이트를 접속하여 확인할수 있다. 

대표적으로, 크롬 개발자 도구를 사용하면, 웹페이지나 사이트에 이용된 코드들을 확인할수도 있다 


# 요청과 응답

Client 와 Server 사이드가 존재하는데. 
Client 는 특정 Server 에 어떠한것을 요청하여, 응답으로 그 값을 받는다. 

예를들어, www.naver.com 이라고, 브라우저 주소창에 주소를 입력하면, 수많은 호스트 중에 네이버의 코드를 가진 호스트를 찾고, 
그 호스트에 요청 (request) 를 보내게 된다. request 를 받은 호스트 즉, 서버는 응답 (response) 로 코드를 반환하고. 
이 반환된 코드는, 사용자의 브라우저에서 해석되어, 브라우저에서 출력 된다. 





# Frontend 와 Backend 


## Frontend 


frontend 는 클라이언트가 직접 보게 되는 화면을 구성하는 영역으로. Frontend 개발자는, 사용자가 직접 보게될 페이지의 화면 구성등을 제작한다. 

대표적으로, 웹페이지의 구조를 담당하고, 틀을 만들어 가는 HTML (Hyper text Markup language) 와 
디자인 효과를 주는, CSS (Cascading Style Sheets) 그리고, 동적으로 다양한 효과를 제공하는 JS (JavaScript) 를 통한 프로그래밍으로 

멋지고, 아름다운 화면, 그리고 동적효과를 사용자에게 보여줄수 있다. 

웹에서의 UI 와 UX 가 부각되면서, frontend 개발에 대한 중요성이 조명받고 있고. 

2018년에 애플이 웹 접근성과 웹 표준에 대해 소송을 당하면서, 프론트앤드 개발에 대한 중요성이 점점 커지고 있다. 


## Backend 


backend 는 직접 사용자들에게 보이지는 않지만, 웹페이지나 사이트 뒤쪽에서, 논리적으로 어떠한 작동을 처리하는것을 담당하고. 
웹 사이트의 전체적인 동작구조를 만들어 가는것이 백앤드이다. 

백앤드는, 웹사이트가 사용자에게 빠르게 반응하도록 효율적인 코드와 알고리즘을 생각하고 구현되어야 하고, 
개발자가 프레임워크를 사용한다면, 해당 프레임워크의 장점들을 살려서 개발할수 있어야 한다. 
또한, 보안상의 이슈로, 비밀번호와 주민등록번호같은 개인정보들은, 암호화 하는 과정도 있어야 한다. 

직접적으로 눈에 보이지는 않지만, 아주 복잡한 기능과 동작들을 수행하는 경우도 많다. 

또한, 시간이 흐르면서, 아무리 잘 작성된 웹사이트라고 하더라도, 유지 보수가 계속 유지되어야 하며, 추가적인 기능을 개발해야하는 경우가 매우 많다. 

따라서, 다른 사람들이 유지보수를 하거나, 추가 기능을 개발할때에 헷갈리지 않도록, 
자신만의 코드 스타일로만 작성하기 보다는, 어느정도 보편화 되고 규격화된 코드 스타일로 작성하는것이 좋다. 

# Django 


Django 는 Web Framework 중 하나이다. 
Framework 는, 특정 기능들을 개발자가 더 빠르게 구현할수 있도록 미리 만들어진 뼈대, 틀이라고 할수 있다. 

웹사이트를 만들기 위해서 필요한 과정들과 개발요소들을 미리 뼈대 틀에 담아놓은 것이 Django 라고 할수 있다. 

프레임워크는, 여러 종류의 프레임워크가 존재하는데. 언어별로도 다양하고, 같은 언어에서도 다른 프레임워크들이 존재한다. 

- JAVA 로 이루어진 Spring 
- PHP 로 이루어진 Laravel 
- Python 으로 이루어진 Django, Flask 혹은 Tornado 가 존재한다 


Django 로 벡엔드를 개발하고, 프론트앤드에 다른 프레임워크를 사용하여 웹사이트를 제작할수 있다. 
Django 와 주로 사용되는 프론트엔드 프레임워크는, Angular JS 나 React 등이 있다. 

Django 는 확실히 백엔드를 개발하는 프레임워크이고, 
클라이언트에게 요청을 받아서 요청을 처리하고, 응답으로 결과를 주거나 처리결과를 알려줄수 있다. 


Django 의 가장 큰 특성은 아래와 같다. 

- 파이썬으로 이루어져 있다. 
- 다양한 라이브러리 기능이 존재한다. 
- 기본적인 admin 이 제공된다.
